####1. 언어<인공지능 시집>
* **'햇살은 유리창을 뚫고'** 
 * 설명
 
 >마이크로소프트는 2017년 5월 중국에서 서비스 중인 인공지능 기반의 챗봇 ‘샤오이스(Xiaoice)’를 활용해 인공지능이 쓴 시를 모은 시집을 발간했습니다. 샤오이스는 1920년 이후 현대 시인 519명의 작품 수천 편을 스스로 학습해서 1만여 편의 시를 지었고 이 중 일부를 시집으로 엮어 출판한 것이라고 합니다. 시집 제목은 ‘햇살은 유리창을 잃고(Sunshine Misses Windows)’로 제목도 샤오이스가 직접 지었다고 합니다.
 
 * 원리

  > 인공지능 창작 작업에 사용되는 대표적인 창작 알고리즘은 *'적대적 생성 신경망(GAN·Generative Adversary Network)'*입니다. 

  > GAN은 생성기와 판별기로 구성된 서로 다른 주체가 적대적으로 경쟁하며 자신의
성능을 강화하는 과정을 통해 진본 데이터에 가까운 위조 데이터를 생성하는 원리입니다. 즉 생성기에서는 임의의 분포로부터 위조 데이터를 생성하고, 판별기는 진본 데이터와 위조 데이터를 구별하기 위해 학습을 진행합니다. 이 과정에서 생성기를 판별기를 최대한 잘 속이기 위해 노력하고 판별기는 진본 데이터와 위조 데이터를 최대한 구별하기 위해 경쟁적으로 학습합니다. 결과적으로 GAN은 진짜와 같아지는 학습을 통해 사용자가 입력한 조건에 가장 가까운 샘플을 만들어 보다 생생한 데이터(이미지, 음성 등)를 생성할 수 있게 됩니다.

 * 현황과 전망

  > 현재 200 여개에 달하는 GAN 변형 기술은 딥러닝이 안고 있는 비교사학습과
라벨링되지 않은 데이터 활용 문제를 해결할 수 있는 돌파구로 인식하고 있습니다.
GAN은 스스로 새로운 지식과 경험을 축적할 수 있는 방향으로 진화할 수 있는
가능성을 확인시켜 주었고 향후 산업적 활용 가능성이 매우 클 것으로 기대되고 있습니다. 그러나 GAN을 이용한 딥페이크(Deep Fake) 등 가짜 콘텐츠 생성으로 인한 심각한 사회 문제를 발생시킬 것에 대한 부작용 우려 또한 증가하고 있습니다.
  
  > GAN은 데이터 생성이 용이하고 스스로 새로운 모형을 생성할 수 있다는 특별한
장점을 지니고 있으나 생성기와 판별기 간 성능 불균형이 큰 경우 학습이 매우 어렵다는 기술적 한계가 있습니다. GAN은 최소최대화(minimax) 문제를 기반으로 두 네트워크를 경쟁적으로 학습시켜 새로운 데이터를 생성하는 데 우수하나 최소최대화의 근본적인 문제가 존재합니다. 이를 해결하기 위한 다양한 기법이 제시되고 있으나 아직까지 생성적 적대 문제를 완벽하게 해결할 수 있는 방안은 없으며 지속적인 연구가 필요합니다.

####2. 음성<인공지능 스피커>
* **구글 홈** 
 * 설명
 
 >구글 홈은 구글에서 만든 스마트 스피커입니다. 스마트 스피커란 IoT 음성인식 인공지능 비서가 탑재된 스피커 및 그 서비스를 총칭하는 용어입니다. 
 
  * 원리

  > 인간의 목소리가 AI 스피커의 마이크를 통해 입력되면 ‘스피치 프로세서(SPEECH PROCESSOR)'는 음성 신호를 음성 신호 벡터로 변화시킵니다. 음성 신호 벡터는 *‘스피치 인식기(SPEECH RECOGNIZER)’*와 *‘언어 특징 추출기(LINGUISTIC FEATURE EXTRACTOR)’*를 통해 *‘언어 표현 정보(LINGUISTIC REPRESENTATION)’*로 추출됩니다. 여기서 언어 표현 정보는 기존의 단어 관련 정보를 고려하여 추출됩니다.
  이렇게 추출된 언어 표현 정보는 ‘스킬 분류기(SKILL CLASSIFIER)’, ‘의도 분류기(INTENT CLASSIFIER)’ 및 ‘슬롯 인지기(SLOT RECOGNIZER)’로 입력됩니다.

  > 먼저 스킬 분류기는 언어 표현 정보를 통해 입력된 명령에 알맞은 스킬을 결정할 수 있습니다. 의도 분류기는 언어 표현 정보를 통해 입력된 명령의 의도를 결정할 수 있습니다.슬롯 인지기는 언어 표현 정보에서 의미 있는 정보를 인지합니다. 이렇게 분류 및 인지를 마친 스킬, 의도, 슬롯은 ‘스킬 프로세서(SKILL PROCESSOR)’를 거쳐 명령 수행을 위한 관련 기기를 동작시킵니다.

 * 현황과 전망

  > 미국의 정보 기술 연구 및 자문 회사인 가트너는 AI 스피커 시장 규모가 2015년 약 4050억 원에서 매년 40% 이상 성장을 예상했고, 2020년에는 약 2조 3600억 원이 될 것으로 전망했습니다. 시장조사기관 트랙티카(Tractica)는 2025년 세계 인공지능 산업 연 매출이 348억 달러(약 43조원)에 달할 것으로 예상하고 있습니다. 시장조사기관 오범(Ovum)은 가상 비서와 음성 AI 를 지원하는 스마트 홈 제품, TV, 웨어러블 기기의 수는 2021 년에 16억 3,000만 대에 육박할 것으로 예상했습니다.

####3. 이미지<사진확대 앱>
* **Resize 2X** 
 * 설명
 
 >이 앱은 이미지를 두배 사이즈로 미세하게 조정하여 늘려줍니다. 이미지를 두배크기로 늘려주지만 흐릿함과 노이즈, 픽셀화, 깨짐과 같은 현상은 최소화 하였습니다. Resize 2X는 이미지의 크기 조정에 있어 최고의 품질을 위해 향상된 AI 알고리즘을 사용하고 있습니다.
 
  * 원리

  > 이 앱은 모델이 직접 이미지, 비디오, 텍스트 또는 사운드를 분류하는 *CNN(convolutional neural network)* 알고리즘을 사용하고 있습니다. 이 알고리즘은 
컨벌루션 뉴럴 네트워크를 구성하는 수십, 수백 개의 계층이 각각 이미지의 서로 다른 특징을 감지하도록 학습할 수 있습니다. 필터는 각 학습 이미지에 서로 다른 해상도로 적용되고, 필터의 출력은 다음 계층의 입력으로 활용됩니다. 필터는 밝기 및 가장자리 등과 같이 매우 단순한 특징에서 시작하여 객체만의 고유한 특징으로 더 복잡하게 발전할 수 있습니다. CNN은 이미지, 텍스트, 사운드, 비디오의 특징을 식별하고 분류합니다.

 * 현황과 전망

  > 객체 검출 및 인식 분야를 발전시키는 CNN. 
CNN은 이미지 인식과 패턴 감지를 위한 최적의 아키텍처를 제공합니다. CNN은 발전을 거듭해 온 GPU와 병렬 컴퓨팅과 함께 활용되어 자율 주행과 얼굴 인식 기술이 새롭게 등장하는 발판을 마련했습니다. 컨벌루션 뉴럴 네트워크는 수백, 수천 개, 때로는 수백만 개의 이미지를 학습합니다. 

  > 다량의 데이터와 복잡한 네트워크 아키텍처가 사용될 때 GPU를 적용하면 모델을 학습시키기 위한 처리 시간을 대폭 단축할 수 있습니다. CNN 학습이 완료되면 ADAS(고급 운전자 지원 시스템)에서의 보행자 감지 등과 같은 실시간 응용 분야에 활용할 수 있습니다.

####4. 자율주행<자율주행 자동차>
* **테슬라 모델3** 
  * 설명
 
 >테슬라 모델3는 오토파일럿을 사용하여 차량을 차선 내에서 자동 조향하거나 가속 및 제동할 수 있습니다. 고속도로 진입부터 진출까지 능동적으로 안내하고, 차량 자동 주차 및 호출이 가능하며, 터치 한번으로 평행 주차 및 직각주차가 가능합니다. 또한 고속도로 주행 중 자동 차선 변경 기능이 있습니다.
 
  * 원리

  > 테슬라 모델3는 *SLAM(Simultaneous localization and mapping)*라는 기법을 사용합니다. 동시적 위치추정 및 지도작성으로 번역되는데 로봇에 의한 지도작성의 한 분야로서, 이름처럼 자신이 알지 못하는 주위 공간의 지도를 만들면서 동시에 자신이 그 지도상의 어디에 있는지를 추정하는 기법입니다. 차량에 붙은 여러 개의 센서와 오프라인 지도를 결합해서 활용합니다. 2016년에 개발된 실시간 객체 인식 알고리즘(Object Detection Algorithm)인 YOLO(You only look once)와 같은 딥러닝 인공지능에 기반한 객체 탐지, 컴퓨터 비전이 사용됩니다.

 * 현황과 전망

  > 현재 많은 자동차 기업뿐만 아니라 구글, NVIDIA 같은 IT 기업들이 개발 중인데, 그중 가장 앞서있다고 평가되는 구글의 자율주행차는 약 600 ~ 1000여대의 차량으로 실제 도로 주행을 하여 주행거리가 1,300만km를 넘어습니다. 이에 반해 테슬라의 경우, 2020년 1월 자율주행 시스템으로 실도로 총주행거리는 30억 km가 넘으며, 70여만대의 자율주행 하드웨어 차량을 통해서 단, 하루에도 약 881만 km의 데이터를 수집할 수 있어서, 데이터의 양이 얼마나 큰 지가 압도적인 차이를 만들어 내는 딥러닝 기반의 자율 주행 개발에 있어서 가장 유리한 위치를 차지하고 있습니다.애플도 장기 프로젝트 중 하나로 자율주행차 개발을 진행하고 있었으나 최근에 관련 인력을 해고하거나 재배치하였습니다. 벤츠, 아우디, 현대자동차, 포드, GM 등 기존 자동차 브랜드들도 개발 중입니다.

  > 현재도 다수의 자동차에서 볼 수 있는 크루즈 컨트롤, 차선이탈 경보 시스템, 주차 보조 시스템 등도 기본적인 자율주행 기술의 하나라고 볼 수 있습니다.
